## Dataset Description

<p>In this project we will primarily reference datasets concerning the NBA. By the end of this project, we wil be able to predict the win/loss ratio of every team in the NBA along with a grade that we will give to their team based on player stats that they have averaged for every season that they have been in the NBA.<br>Using Basketball Reference, and more specifically https://www.basketball-reference.com/players/ for every player that we will consider (active players only), and https://www.basketball-reference.com/leagues/ for the previous records that teams achieved. <br>For training, we will consider the stats that players have previously achieved (ie. their averages for points, assists, rebounds, steals, blocks, fouls etc.) and compile that into team stats. (We cannot use base team stats for testing as future NBA teams will not be composed of the same players that they currently have.) Using this trainging set we can also train the win loss categor for teams based on these stats and then use similar testing sets to make predictions about the future, such as the 2025-2026 season and 2026-2027 season.</p>

## Technical Implementation

<p>As this project will be much larger and wider in scope from previous programming assignments that we have completed in this class, we expect there to be a lot of difficulties in pre-processing<br><br>A large one that we are currently considering is the fact that some players may have spenta lot of time hurt in their carrer and may be missing some stats in some categories or will even be considered in active going into seasons or may be on Free Agency and will not yet be associated with a team or may even be traded to other teams which could affect the outcome of our predictions throughout different seasons. To combat this, in training, we will make it a goal to only consider teams' stats after seasons have ended. <br><br>Another aspect we are considering is the overall development of players. Although not a direct issue with technical implementation, younger players are often inconsistent in their earkly years and may see a lot of growth or decline depending on the time that they are allowed to play. <br><br>This may cause issues with our training end testing when we attempt to build a classifier that is accurate, so we have a number of options to consider when dealing with cases like this. Voiding younger players stats, or not counting them as much when testing could be possible options. We may also have to consider rookies that will be playing for teams in the future because they do not have any stats in the NBA. In this case, we may want to create a small classifier to predict a few stats based on college performance which is data we can alos obtain on Basketball Reference. </p>
